```python
column=df.columns
column
```

```python
#Compare the distribution of various Parameters with respect to Class 0 and 1 with the help of pairplot
for var in column:
    sns.pairplot(df, x_vars=var,y_vars='Class',size=4,aspect=0.7)
```

```python
#from the above pairplots we can observe that for class 1 amount is comparitively very less than those of th class 0.
sns.pairplot(df, x_vars='Amount',y_vars='Class',size=7,aspect=0.7)
```

```python
sns.pairplot(df, x_vars='Time',y_vars='Class',size=7,aspect=0.7)
```

```python
from matplotlib.colors import ListedColormap
plt.figure(figsize=(20,18))
#corr=train_dataset.corr()
corr=df.astype(float).corr()
sns.set(style='ticks',color_codes=True)
sns.heatmap(corr,linewidths=0.3,square=True,linecolor='white',annot=True)
```

```python
corr_list=corr['Class'].sort_values(axis=0,ascending=False).iloc[1:]
corr_list
```
